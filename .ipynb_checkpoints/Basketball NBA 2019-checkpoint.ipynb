```python
import pandas as pd

url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'

html = pd.read_html(url , header= 0)

df_2019 = html[0]

```

```python
df_2019["Age"].unique()
```

```python
#lets drop age

raw = df_2019.drop(df_2019[df_2019["Age"] == "Age" ].index )
```

```python
#check
raw["Age"].unique()
```

```python
raw["Rk"].isnull().sum()
```

```python
#dropping rk column
df = raw.drop(["Rk"] , axis = 1)
df
```

```python
#checking the null values in each column

df.isnull().sum()
```

```python
#fill null values with 0

df = df.fillna(0)
df.isnull().sum()
```

```python
#converting and saving dataset as csv

df.to_csv("NBA_2019_per_game" , index = False)
```

# GroupBy() function

```python
lal = new_df.groupby("Tm").get_group("LAL")
max_lal = lal[lal["PTS"] == lal["PTS"].max()]
```

```python
# Of the 5 positions, which position scores the most points?

# We first group players by their positions.


new_df["Pos"].unique()
```

```python
#We will now show only the 5 traditional positions (those having combo positions will be removed from the analysis).
positions = ["SG" , "PF", "PG","C","SF"]

top_POS = new_df[new_df["Pos"].isin(positions)]
```

```python
# We first group players by their positions.

top_POS.groupby("Pos")["PTS"].describe()
```

```python
# heat map
## compute coorelation matrix

cdf = new_df.corr()
cdf
```

# make the heat map

```python
sns.heatmap(cdf)
```

```python
# Adjust figure size of heat map

import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(cdf, square=True)
```

# Mask diagonal half of heat map (Diagonal correlation matrix)

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html

import numpy as np
import seaborn as sns

mask = np.zeros_like(cdf)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(cdf, mask=mask, vmax=1, square=True)
```
